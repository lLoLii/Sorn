```python
def take_semester_df(lst_subj_NAF=None,lst_subj_AF=None,except_NAF=[],except_AF=[],name = None,semester = '6/2',data = None):


    
    if lst_subj_NAF == 'all':
        lst_subj_NAF = ['phy60','chem60','bio60','social','th','eng','he','pe','art','mathex','astro']


    if lst_subj_AF == 'all':
        lst_subj_AF = ['phy60','chem60','bio60','social','th','eng','he','pe','art','mathex','astro']

    if lst_subj_AF == 'allterm21':
        lst_subj_AF = ['math','phy60','chem60','social','th','eng','he','pe','mathex','astro','bio60']


    if lst_subj_NAF == 'allterm2':
        lst_subj_NAF = ['math','phy60','chem60','social','th','eng','he','pe','mathex','astro','bio60']

    if lst_subj_NAF == 'allterm21':
        lst_subj_NAF = ['math','phy60','chem60','social','th','eng','he','pe','mathex','astro','bio60']


    ind = df[(df['name']== name) & (df['semester']== semester)].index[0]
    al_sbj_done = list(data[data['name']== name ]['subject_done'].values)[0]
    al_sbj_redo = list(data[data['name']== name ]['subject_redo'].values)[0]
    
    if type(al_sbj_done) == type(1.0):
        al_sbj_done = []
        except_NAF = []
    else:
        except_NAF = al_sbj_done.replace("'",'').replace("[",'').replace("]",'').replace(" ",'').split(',')

    if (al_sbj_redo == []) or(al_sbj_redo == '[]') or (type(al_sbj_redo) == type(1.0)) or (type(al_sbj_redo) == np.float64):
        al_sbj_redo = []
        lst_subj_AF = []
    else:
        lst_subj_AF = al_sbj_redo.replace("'",'').replace("[",'').replace("]",'').replace(" ",'').split(',')
    

    if not except_NAF == []:
        for rm_subj in except_NAF:
            lst_subj_NAF.remove(rm_subj)

        msg = listToString(except_NAF) + ' REMOVED'
        time.sleep(2)
        notify(msg)
        print(msg)
    else: 
        except_NAF = []

    notify('except_NAF = ' + str(except_NAF))


    
    if not except_AF == None:
        for rm_subj in except_AF:
            lst_subj_AF.remove(rm_subj)

        msg = listToString(except_AF) + ' REMOVED'
        time.sleep(2)
        notify(msg)
        print(msg)

    if not lst_subj_NAF == None:
        random.shuffle(lst_subj_NAF)
        subj_todo = len(lst_subj_NAF)
        subj_done = 0
        while not subj_done == subj_todo:
            for subj in lst_subj_NAF:
                if subj == 'eng': lang = 'en'
                else: lang = 'th'
                pag.moveTo(100,100)
                time.sleep(2)
                click_SUBJ(subj)

                d = take_subject(False,lang,1,subj,name)
                if d == False:
                    msg = '\nTaking {} in Semester 1'.format(subj)
                    click_semester1()
                    click_SUBJ(subj)
                    take_subject(False,lang,1,subj)
                    click_HOME()
                    time.sleep(0.8)
                    pag.click(1210,875)
                    time.sleep(0.8)
                    pag.click(1210,875)
                    time.sleep(0.8)
                    pag.click(1210,875)
                    click_SUBJ(subj)
                    take_subject(False,lang,1,subj)


                msg = subj + ' DONE'
                notify(msg)
                print(msg)
                except_NAF.append(subj)
                df.at[ind,'subject_done'] = except_NAF
                df.to_csv(user_data_path)
                file2.SetContentFile(user_data_path)
                file2.Upload()
                subj_done = subj_done + 1
                time.sleep(2)


                click_CHANGESUBJ()


    if not lst_subj_AF == None:
        random.shuffle(lst_subj_AF)
        subj_todo = len(lst_subj_AF)
        subj_done = 0
        while not subj_done == subj_todo:
            for subj in lst_subj_AF:
                if subj == 'eng': lang = 'en'
                else: lang = 'th'
                pag.moveTo(100,100)
                time.sleep(2)
                click_SUBJ(subj)

                take_subject(True,lang,5,subj,name)

                msg = subj + ' DONE'
                notify(msg)
                print(msg)
                subj_done = subj_done + 1
                lst_subj_AF.remove(subj)
                df.at[ind,'subject_redo'] = lst_subj_AF
                df.to_csv(user_data_path)
                file2.SetContentFile(user_data_path)
                file2.Upload()

                msg = listToString(subj) + 'REDO REMOVED'
                notify(msg)

                time.sleep(2)

                click_CHANGESUBJ()
            
```
